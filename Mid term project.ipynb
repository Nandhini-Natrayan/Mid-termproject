```python
import pandas as pd
from itertools import chain

# read the CSV file into a pandas dataframe
df = pd.read_excel('usedCarsFinal.xlsx')
df = df.dropna(subset=['other_features_list'])

df
```

```python

cat_col = df['other_features_list'].dropna(axis=0)

# split each row into separate categorical values
cat_list = [x.split(',') for x in cat_col]

# flatten the list of lists into a single list
flat_cat_list = list(chain(*cat_list))

# extract the unique categorical values
unique_cat_values = set(flat_cat_list)

# print the unique categorical values
a=(unique_cat_values)
a
```

```python
unique_cats = set()
for row in cat_col:
    cats = row.split(',')
    for cat in cats:
        unique_cats.add(cat.strip())
```

```python
for cat in unique_cats:
    df[cat] = 0
print(df)    
```

```python
for i, row in df.iterrows():
    cats = row['other_features_list'].split(',')
    for cat in cats:
        cat = cat.strip()
        if cat in unique_cats:
            df.at[i, cat] = 1
```

```python
def rename_columns(df):
    df.rename(columns={'mileage':'mileage(km)',
                     'Immobilizer Key':'immobilizer_key',
                     'engine_capacity': 'engine_capacity(cc)',
                     'USB and Auxillary Cable' : 'USB_and_Auxillary_Cable',
                     'CD Player' : 'CD_Player',
                     'AM/FM Radio' : 'AM/FM_Radio',
                     'Heated Seats' : 'Heated_Seats',
                     'Alloy Rims' :  'Alloy_Rims',
                     'Rear Camera' : 'Rear_Camera',
                     'Power Steering' :  'Power_Steering',
                     'Sun Roof' : 'Sun_Roof',
                     'DVD Player': 'DVD_Player',
                     'Cruise Control' : 'Cruise_Control',
                     'Power Windows' : 'Power_Windows',
                     'Climate Control': 'Climate_Control',
                     'Front Camera' : 'Front_Camera',
                     'Power Locks':  'Power_Locks',
                     'Air Conditioning':'Air Conditioning',
                     'Navigation System' : 'Navigation_System',
                     'Steering Switches' :  'Steering_Switches',
                     'Power Mirrors' : 'Power_Mirrors',
                     'Rear AC Vents':  'Rear_AC_Vents',
                     'Keyless Entry' : 'Keyless_Entry',
                     'Front Speakers' : 'Front_Speakers',
                     'Air Bags' : 'Air_Bags',
                     'Rear Seat Entertainment' :'Rear_Seat_Entertainment',
                     'Cassette Player': 'Cassette_Player',
                     'Rear Speakers' : 'Rear_Speakers'}, inplace=True )
    return df
```
